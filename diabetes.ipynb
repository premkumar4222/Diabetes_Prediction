<h4>Importing Libraries</h4>

```python
import pandas as pd
import seaborn as sns
```

<h4>Importing DataSet</h4>

```python
data=pd.read_csv('diabetes.csv')
```

```python
data.head()
```

```python
data.info()
```

```python
#Checking for missing values
data.isnull().sum()
```

```python
#There is no missing values
```

```python
data.shape
```

```python
# splitting data into predictor variables and target 
X=data.drop('Outcome',axis=1)
y=data['Outcome']
```

```python
X.head()
```

```python
y.head()
```

<h4>Splitting Data into Train and Test </h4>

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train,y_test = train_test_split(X,y ,train_size=0.8)
```

```python
X_train.head()
```

<h4>Training Model</h4>

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
model = DecisionTreeClassifier()
```

```python
model.fit(X_train,y_train)
```

```python
y_pred=model.predict(X_test)
```

```python
y_pred
```

<h4>Evaluating Model</h4>

```python
from sklearn import metrics
```

```python
print(metrics.accuracy_score(y_test,y_pred))
```

```python
# We got accuracy only 67% so now we are trying to improve accuracy by using hyper parameter tuning
```

<h4> Applying Hyper Parameter Tuning </h4>

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid = {
    'max_depth': [3, 5, 7, 10],
    'min_samples_leaf': [1, 5, 10, 20],
    'criterion': ['gini', 'entropy']
}
```

```python
dt_classifier = DecisionTreeClassifier()

# Perform grid search with cross-validation
grid_search = GridSearchCV(estimator=dt_classifier, param_grid=param_grid, cv=5, scoring='accuracy')
grid_search.fit(X_train, y_train)
```

```python
print("Best Hyperparameters:", grid_search.best_params_)
```

```python
best_dt_classifier = DecisionTreeClassifier(**grid_search.best_params_)
best_dt_classifier.fit(X_train, y_train)
```

```python
y_pred = best_dt_classifier.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test, y_pred)
print("Test Set Accuracy:", accuracy)
```
